```python
import hashlib
import time

# ---------------- BLOCK CLASS ----------------
class Block:
    def __init__(self, index, voter_id, candidate_id, previous_hash):
        self.index = index
        self.timestamp = time.time()
        self.voter_id = voter_id
        self.candidate_id = candidate_id
        self.previous_hash = previous_hash
        self.hash = self.calculate_hash()

    def calculate_hash(self):
        data = str(self.index) + str(self.timestamp) + str(self.voter_id) + str(self.candidate_id) + self.previous_hash
        return hashlib.sha256(data.encode()).hexdigest()


# ---------------- BLOCKCHAIN CLASS ----------------
class Blockchain:
    def __init__(self):
        self.chain = [self.create_genesis_block()]

    def create_genesis_block(self):
        return Block(0, "None", "None", "0")

    def add_block(self, voter_id, candidate_id):
        prev_block = self.chain[-1]
        new_block = Block(len(self.chain), voter_id, candidate_id, prev_block.hash)
        self.chain.append(new_block)

    def print_chain(self):
        for block in self.chain:
            print("\n-------------------------")
            print(f"Index: {block.index}")
            print(f"Voter ID: {block.voter_id}")
            print(f"Candidate ID: {block.candidate_id}")
            print(f"Hash: {block.hash}")
            print(f"Previous Hash: {block.previous_hash}")

    def is_valid(self):
        for i in range(1, len(self.chain)):
            curr = self.chain[i]
            prev = self.chain[i-1]

            if curr.hash != curr.calculate_hash():
                return False

            if curr.previous_hash != prev.hash:
                return False
        return True


# ---------------- MAIN SYSTEM ----------------
class VotingSystem:
    def __init__(self):
        self.voters = {}
        self.candidates = {}
        self.blockchain = Blockchain()

    def add_voter(self):
        vid = input("Enter Voter ID: ")
        if vid in self.voters:
            print("❌ Voter already exists!")
            return
        name = input("Enter Voter Name: ")
        self.voters[vid] = {"name": name, "has_voted": False}
        print("✅ Voter added successfully!")

    def add_candidate(self):
        cid = input("Enter Candidate ID: ")
        if cid in self.candidates:
            print("❌ Candidate already exists!")
            return
        name = input("Enter Candidate Name: ")
        self.candidates[cid] = name
        print("✅ Candidate added successfully!")

    def cast_vote(self):
        vid = input("Enter Voter ID: ")
        if vid not in self.voters:
            print("❌ Voter not found!")
            return

        if self.voters[vid]["has_voted"]:
            print("❌ Already voted!")
            return

        cid = input("Enter Candidate ID: ")
        if cid not in self.candidates:
            print("❌ Candidate not found!")
            return

        self.blockchain.add_block(vid, cid)
        self.voters[vid]["has_voted"] = True
        print("✅ Vote cast successfully!")

    def print_blockchain(self):
        self.blockchain.print_chain()

    def validate_chain(self):
        if self.blockchain.is_valid():
            print("✅ Blockchain is VALID")
        else:
            print("❌ Blockchain is CORRUPTED")


# ---------------- MENU ----------------
def main():
    system = VotingSystem()

    while True:
        print("\n====== Voting System Menu ======")
        print("1. Add Candidate")
        print("2. Add Voter")
        print("3. Cast Vote")
        print("4. Print Blockchain")
        print("5. Validate Chain")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            system.add_candidate()
        elif choice == "2":
            system.add_voter()
        elif choice == "3":
            system.cast_vote()
        elif choice == "4":
            system.print_blockchain()
        elif choice == "5":
            system.validate_chain()
        elif choice == "6":
            print("Exiting...")
            break
        else:
            print("❌ Invalid choice!")

if __name__ == "__main__":
    main()
```
